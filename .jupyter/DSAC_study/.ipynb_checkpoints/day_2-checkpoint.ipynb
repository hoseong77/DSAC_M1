### 몬테카를로 분석
random 숫자 발생을 이용하여 원주율 pi를 시뮬레이션을 구하는 방법
원의 면적: pi * r ** 2

pi=4*(random한 점이 원내부에 들어올 활률)

10,000개의 난수 순서쌍 (x, y)(x,y) 가운데에는 x^2 + y^2 <=1을 만족하여 원 안의 범위에 포함되는 것들이 있을 것이다.   
그런 순서쌍들의 개수를 세어 전체 난수에 대한 비율을 계산하면 대략적인 원의 면적을 구할 수 있다. 무작위로 뽑힌 난수의 개수가 늘어날수록 더 정확한 결과를 얻을 수 있으나, 그만큼 더 많은 시간이 걸리는 것을 감안


```python
n = 10000
x = np.random.rand(n).round(2) 
y = np.random.rand(n).round(2) 
x,y
```

```python
plt.figure(figsize=(6,6)) 
plt.scatter(x,y, s=3)
```

```python
xx = np.linspace(0,1,100)
plt.plot(xx, (1-xx*xx)**0.5, c = 'r')  #y2= 루트 1-x2
```

```python
z = np.random.rand(500000)
(z<0.5).mean()
```

```python
n = 10000
x = np.random.rand(n)
y = np.random.rand(n)

plt.figure(figsize=(6,6)) 
plt.scatter(x,y, s=1)

xx = np.linspace(0,1,100)
plt.plot(xx, (1-xx*xx)**0.5, c = 'r')  #y2= 루트 1-x2

pi = ((x**2+y**2)<1).mean()*4
print(pi)

print(sum((x**2+y**2)<1)*4/n)
```

### 람다 함수

-람다함수를 통해 이름이 없는 함수를 만들 수 있습니다.

-람다함수의 장점은 코드의 간결함 메모리의 절약이라고 생각할 수 있습니다.

lambda x : x + 1


```python
fc = lambda x : x + 1

fc(10)
```

```python
f = lambda x : x.max() - x.min()

df = pd.DataFrame(np.arange(0,12,1).reshape(4,3),
                  columns=["A","B","C"],index = ["a","b","c","d"])
df
```

```python
df.apply(f,0)
```

```python
df.apply(f,1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame, Series
from numpy.random import randn
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
ax1.hist(np.random.randn(200), bins = 10, color ="r", rwidth = 0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3*np.random.randn(30))
ax3.plot(np.arange(10),np.random.randn(10))

fig

#코드를 두번 실행하면 계속 겹쳐서 그린다  해결 방법은 위 코드 실행 후 다시 실행
```

```python
axes = plt.subplots(2,3)
```

```python
plt.subplots_adjust(left=None, bottom=None, right=None, top=None,
                wspace=None, hspace=None)

# subplot 간의 간격을 조정하는 함수 _adjust()
# wspace, hspace: subplot 간의 너비 높이
```

```python
fig, axes = plt.subplots(2,2, sharex = True, sharey = True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)

plt.subplots_adjust(wspace=0,hspace=0)

# sharex : 모든 그래프에서 x 값의 범위가 동일한지 여부 
# sharey : 모든 그래프에서 y 값의 범위가 동일한지 여부
```

```python
plt.plot(randn(30), "ko--") #ko--, k^, k:
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
ax1.plot(randn(1000).cumsum(),"k")
ax2.plot(randn(1000).cumsum(),"b--")
ax3.plot(randn(1000).cumsum(),"r^")
ax4.plot(randn(1000).cumsum(),"go")
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(),"k",label = "one")
ax.plot(randn(1000).cumsum(),"b--",label = "two")
ax.plot(randn(1000).cumsum(),"r",label = "three")
ax.plot(randn(1000).cumsum(),"g",label = "four")

ax.legend(loc = "best")
```
